# 1. Support Vector Machine

## 1.1 Dataset 1

```python
data1 = loadmat("ex6data1.mat")
```

```python
X1 = data1['X']
y1 = data1['y']

df1 = pd.DataFrame(X1, columns=['X1', 'X2'])
df1['y'] = pd.DataFrame(y1)
```

```python
positive1 = df1[df1['y'] == 1]
negative1 = df1[df1['y'] == 0]
fig, ax = plt.subplots()
ax.scatter(x=positive1.X1, y=positive1.X2, marker='X', label='Positive')
ax.scatter(x=negative1.X1, y=negative1.X2, marker='o', c='r', label='Negative')
ax.legend()
plt.show()
```

```python
model1 = svm.LinearSVC(C=100, loss='hinge')
model1.fit(X1,y1.ravel())
print(model1.score(df1[['X1', 'X2']], df1['y']))

fig, ax = plt.subplots()
ax.scatter(x=positive1.X1, y=positive1.X2, marker='X', label='Positive')
ax.scatter(x=negative1.X1, y=negative1.X2, marker='o', c='r', label='Negative')
ax.legend()

xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 50)
yy = np.linspace(ylim[0], ylim[1], 50)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = model1.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors=['y', 'g', 'y'],levels=[-1,0,1], alpha=1)

plt.show()
```

## 1.2 SVM With Gaussian Kernels

```python
def gaussianKernel(sigma, x1, x2):
    top = np.sum(np.power(x1-x2, 2))
    bottom = 2*np.power(sigma, 2)
    return np.exp(-(top/bottom))
```

### 1.2.2 Dataset 2

```python
data2 = loadmat("ex6data2")
df2 = pd.DataFrame(data2['X'], columns=['X1','X2'])
df2['y'] = pd.DataFrame(data2['y'], columns=['y'])
positive2 = df2[df2['y'] == 1]
negative2 = df2[df2['y'] == 0]
fig, ax = plt.subplots()
ax.scatter(x=positive2.X1, y=positive2.X2, marker='X', label='Positive')
ax.scatter(x=negative2.X1, y=negative2.X2, marker='o', c='r', label='Negative')
ax.legend()

plt.show()
```

```python
import itertools

C = [0.05, 0.1, 1, 10, 100, 1000]
Gamma = [0.1, 1, 10, 100]
maxScore = 0
maxC = 0
maxG = 0
for c,gamma in itertools.product(C, Gamma):
    model2 = svm.SVC(kernel='rbf', C=c, gamma=gamma, probability=True)
    model2.fit(df2[['X1','X2']], df2['y'])
    score = model2.score(df2[['X1', 'X2']], df2['y'])
    if score > maxScore:
        maxC = c
        maxG = gamma
        maxScore = score

model2 = svm.SVC(kernel='rbf', C=maxC, gamma=maxG, probability=True)
model2.fit(df2[['X1','X2']], df2['y'])


fig, ax2 = plt.subplots()
ax2.scatter(x=positive2.X1, y=positive2.X2, marker='X', label='Positive')
ax2.scatter(x=negative2.X1, y=negative2.X2, marker='o', c='r', label='Negative')

xlim1 = ax2.get_xlim()
ylim1 = ax2.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim1[0], xlim1[1], 30)
yy = np.linspace(ylim1[0], ylim1[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.c_[XX.ravel(), YY.ravel()]
Z = model2.predict(xy).reshape(XX.shape)
ax2.contour(XX, YY, Z, alpha=1)
plt.show()
```

## 2. Spam Classification

```python
data_train = loadmat("spamTrain.mat")
data_test = loadmat("spamTest.mat")
Xtrain, ytrain, Xtest, ytest = data_train['X'], data_train['y'].ravel(), data_test['Xtest'], data_test['ytest'].ravel()
C = [0.1, 1, 5, 10, 50, 100]
Gamma = [0.1, 1, 5, 10, 50]
maxScore, maxC, maxG = 0, 0, 0

for c, gamma in itertools.product(C, Gamma):
    model = svm.SVC(C=c, gamma=gamma, kernel='rbf')
    model.fit(Xtrain, ytrain)
    score = model.score(Xtest, ytest)
    if score > maxScore:
        maxScore = score
        maxC = c
        maxG = gamma

maxModel = svm.SVC(C=maxC, gamma=maxG)
maxModel.fit(Xtrain, ytrain)
print(maxModel.score(Xtest, ytest))
```
